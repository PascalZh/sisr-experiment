### Prepare

```python
%load_ext autoreload
%autoreload 2
import scipy.signal
import cv2
import matplotlib.pyplot as plt
import si_algo_model
import si_algo_train
si = si_algo_model.SuperInterpolation()
downscale = si_algo_model.Downscale()
```

```python
GT = cv2.imread("images/LenaRGB.bmp")
LR = downscale(GT)
```

### Train

```python
si.fit(si_algo_train.dataloader())
```

```python
train_data = si._train_data_cache
```

### Plot the number of samples in each classes

```python
nums_samples = [d[0].shape[1] for d in train_data]
num_classes = len(train_data)

plt.figure(layout="constrained", dpi=150)
plt.subplot(211)
plt.plot(range(num_classes), nums_samples)
plt.title("All classes")

plt.subplot(212)
except_n = 1
plt.plot(range(except_n, num_classes), nums_samples[except_n:])
plt.title(f"Excepting first {except_n} classes")
plt.savefig("classes_dist.png")
```
